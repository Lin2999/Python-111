```python
def forwari_propagate(X, theta1, theta2):
    m = X.shape[0]
    
    a1 = np.insert(X, 0, values=np.ones(m), axis=1)
    z2 = a1 * theta1.T
    
    a2 = np.insert(sigmid(z2), 0, values=np.ones(m), axis=1)
    z3 = a2 * theta2.T
    h = sigmid(z3)
    
    return a1, z2, a2, z3, h
    
```

```python
def nnCostFunction(nn_params, input_layer_size, hidden_layer_size, num_label, X, y, Lambda):
    length = nn_params.shape[0]
    Theta1 = nn_params[0:hidden_layer_size*(input_layer_size+1)].reshape(hidden_layer_size, input_layer_size+1)
    Theta2 = nn_params[hidden_layer_size*(input_layer_size+1):length].reshape(num_label, hidden_layer_size+1)
    
    m = X.shape[0]
    class_y = np.zeros((m, num_label)) # 数据y对应0-9 映射0/1关系
    for i in range(num_label):
        class_y[:, i] = np.int32(y == i).reshape(1, -1)
    
    # 正则化从1开始 去除一列
    Theta1_colCount = Theta1.shape[1]
    Theta1_x = Theta1[:, 1:Theta1_colCount]
    
    Theta2_colCount = Theta2.shape[1]
    Theta2_x = Theta2[:, 1:Theta2_colCount]
    
    # 正则化向 theta ^ 2
    term = np.dot(np.transpose(np.vstack((Theta1_x.reshape(-1,1),Theta2_x.reshape(-1,1)))),np.vstack((Theta1_x.reshape(-1,1),Theta2_x.reshape(-1,1))))    
    
    # 正向传播
    a1 = np.hstack((np.ones((m, 1)), X))
    z2 = np.dot(a1, np.transpose(Theta1))
    a2 = sigmid(z2)
    a2 = np.hstack((np.ones((m, 1)), a2))
    z3 = np.dot(a2, np.transpose(Theta2))
    h = sigmid(z3)
    
    J = -(np.dot(np.transpose(class_y.reshape(-1, 1)), np.log(h.reshape(-1, 1))) + np.dot(np.transpose(1-class_y.reshape(-1, 1)), np.log(1-h.reshape(-1, 1))) - Lambda * term / 2)/ m
    return np.ravel(J)
    
    
    
```

```python
def randInitializWeights(L_in, L_out):
    W = np.zeros((L_out, 1+L_in))
    epsilon_init = (6.0 / (L_out + L_in)) ** 0.05
    W = np.random.rand(L_out, 1+L_in)*2*epsilon_init - epsilon_init
    return W
```

```python
def backprop(params, input_size, hidden_size, num_label, X, y, learning_rate):
    m = X.shape[0]
    X = np.matrix(X)
    y = np.matrix(y)
     # 重塑
    theta1 = np.matrix(np.reshape(params[:hidden_size * (input_size + 1)], (hidden_size, (input_size + 1))))
    theta2 = np.matrix(np.reshape(params[hidden_size * (input_size + 1):], (num_label, (hidden_size + 1))))
    a1, z2, a2, z3, h = forwari_propagate(X, theta1, theta2) # 正向
    
    J = 0
    # 增加正则项
    delta1 = np.zeros(theta1.shape)
    delta2 = np.zeros(theta2.shape)
    for i in range(m):
        first_term = np.multiply(-y[i, :], np.log(h[i, :]))
        second_term = np.multiply((1 - y[i, :]), np.log(1 - h[i, :]))
        J += np.sum(first_term - second_term)
        
    J = J / m
    J += (float(learning_rate) / (2 * m)) * (np.sum(np.power(theta1[:, 1:], 2)) + np.sum(np.power(theta2[:, 1:], 2)))
    # 反向传播
    for t in range(m):
        alt = a1[t, :]
        z2t = z2[t, :]
        a2t = a2[t, :]
        ht = h[t, :]
        yt = y[t, :]
        
        d3t = ht - yt
        z2t = np.insert(z2t, 0, values=np.ones(1))
        d2t = np.multiply((theta2.T * d3t.T).T, sigmid_gradient(z2t))
        
        delta1 = delta1 + (d2t[:, 1:]).T * alt
        delta2 = delta2 + d3t.T * a2t
        
    delta1 = delta1 / m
    delta2 = delta2 / m
    
    delta1[:, 1:] = delta1[:, 1:] + (theta1[:, 1:] * learning_rate) / m
    delta2[:, 1:] = delta2[:, 1:] + (theta2[:, 1:] * learning_rate) / m
    
    # 梯度矩阵分解
    grad = np.concatenate((np.ravel(delta1), np.ravel(delta2)))
    return J, grad
    
    
    
```
