```python
def plot_init_data(data, fig, ax):
    positive = data[data['y'].isin([1])]
    negative = data[data['y'].isin([0])]
    
    ax.scatter(positive['X1'], positive['X2'], s=50, marker='x', label='Positive')
    ax.scatter(negative['X1'], negative['X2'], s=50, marker='o', label='Negative')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plot_init_data(data, fig, ax)
plt.show()
```

```python
svc = svm.LinearSVC(C=100, loss='hinge', max_iter=1000) # 决策边界
# C loss损失函数 =>hinge 标准   squared_hinge -> 标准的平方
```

```python
svc
```

```python
svc.fit(data[['X1', 'X2']], data['y'])
```

```python
svc.score(data[['X1', 'X2']], data['y'])
```

```python
data.describe()
```

```python
def find_decision(svc, x1imin, x1max, x2min, x2max, diff):
    # 坐标轴刻度线
    x1 = np.linspace(x1imin, x1max, 1000)
    x2 = np.linspace(x2min, x2max, 1000)
    cordinates = [(x, y) for x in x1 for y in x2]
    
    x_cord, y_cord = zip(*cordinates)
    c_val = pd.DataFrame({'x1':x_cord, 'x2':y_cord})
    c_val['cval'] = svc.decision_function(c_val[['x1', 'x2']]) # 最佳分割线
    
    # 决策边界
    decision = c_val[np.abs(c_val['cval']) < diff]
    return decision.x1, decision.x2
```

```python
x1, x2 = find_decision(svc, 0, 4, 1.5, 5, 2 * 10 ** -3)
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x1, x2, s=10, c='r', label='Boundary')
plot_init_data(data, fig, ax)
plt.show()
```

```python
def guassion_kernel(x1, x2, sigma):
    return np.exp(-(np.sum((x1 - x2) ** 2) / (2 * (sigma ** 2))))
```

```python
x1 = np.array([1.0, 2.0, 1.0])
x2 = np.array([0.0, 4.0, -1.0])
sigma = 2
guassion_kernel(x1, x2, sigma)
```

```python
raw_data = loadmat('./ex6data3.mat')
```

```python
raw_data
```

```python
X = raw_data['X']
Xval = raw_data['Xval']
y = raw_data['y'].ravel()
yval = raw_data['yval'].ravel()

fig, ax = plt.subplots(figsize=(12, 8))
data = pd.DataFrame(raw_data.get('X'), columns=['X1', 'X2'])
data['y'] = raw_data.get('y')
plot_init_data(data, fig, ax)
plt.show()
```

```python
C_values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100]
gamma_values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100]

best_score = 0
best_params = {'C':None, 'gamma':None}

for C in C_values:
    for gamma in gamma_values:
        svc = svm.SVC(C=C, gamma=gamma)
        svc.fit(X, y)
        score = svc.score(Xval, yval)
        
        if score > best_score:
            best_score = score
            best_params['C'] = C
            best_params['gamma'] = gamma
```

```python
best_score
```

```python
best_params
```

```python
svc = svm.SVC(C = best_params['C'], gamma=best_params['gamma'])
svc.fit(X, y)
```

```python
def plot_decisionBoundary(X, y, model, class_='linear'):
    plt = plot_data(X, y)
    
    if class_ == 'linear':
        w = model.coef_  # 系数
        b = model.intercept_
        #     x1 = np.linspace(x1imin, x1max, 1000)
        #     x2 = np.linspace(x2min, x2max, 1000)
        xp = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
        yp = -(w[0, 0] * xp + b) / w[0, 1]
        plt.plot(xp, yp, 'b-', linewidth=2.0)
        plt.show()
        print('123')
        
    else:
        x_1 = np.transpose(np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100).reshape(1, -1))
        x_2 = np.transpose(np.linspace(np.min(X[:, 1]), np.max(X[:, 1]), 100).reshape(1, -1))
        X1, X2 = np.meshgrid(x_1, x_2)
        vals = np.zeros(X1.shape)
        for i in range(X1.shape[1]):
            this_X = np.hstack((X1[:, i].reshape(-1, 1), X2[:, i].reshape(-1, 1)))
            vals[:, i] = model.predict(this_X)
        
        plt.contour(X1, X2, vals, [0, 1], color='blue')
        plt.show()
```
